# 1.2 Imputation #

### Step 1: Prepare the data. ###

```python
import pandas as pd
import numpy as np

# import the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
dfx = pd.DataFrame(data=iris.data, columns=iris.feature_names)
c = pd.DataFrame({"sepal length (cm)": [np.nan, 4.9],
                  "sepal width (cm)": [np.nan, 3],
                  "petal length (cm)": [np.nan, 1.4],
                  "petal width (cm)": [0.2, np.nan]})

df = pd.concat([dfx, c], ignore_index=True)
df
```

### Step 2: Impute the missing value with the mode. ###

```python
from sklearn.impute import SimpleImputer
data = SimpleImputer(missing_values=np.nan, strategy="most_frequent")

# Normalize the imputed data.
dataMode = data.fit_transform(df)
dataMode
```

```python
mode = pd.DataFrame(data=dataMode, columns=['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
mode
```

### Step 3: Impute the missing value with the mean ###

```python
data = SimpleImputer(missing_values=np.nan, strategy='mean')
dataMean = data.fit_transform(df)
dataMean
```

```python
mean = pd.DataFrame(data=dataMean, columns=['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
mean
```

### Step 4: Impute the missing value with the median. ###

```python
data = SimpleImputer(missing_values=np.nan, strategy='median')
dataMedian = data.fit_transform(df)
dataMedian
```

```python
median = pd.DataFrame(data=dataMedian, columns=['sepal length (cm)','sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
median
```
